```python
results_1st_order = pd.read_parquet('results/results_opt_first_order_2022-04-06.parquet')
results_2nd_order = pd.read_parquet('results/results_opt_second_order_2022-04-06.parquet')

resultk1 = results_1st_order
resultk2 = results_2nd_order

#result
```

```python
jobsd = {}
target= resk1e1
result= resultk1
for k in keys:
    obs = k
    #try
    order = result.iloc[obs]['order']
    STEPS = result.iloc[obs]['n_steps']
    min_t = result.iloc[obs]['t_min']
    best_params = result.iloc[obs]['best_params']
    print('Order:', order, '| n_steps:', STEPS, '| t_min:',min_t)
    print()
    for c,jbid in enumerate(target[k]):
        jobsd[c] = jakarta.retrieve_job(jbid)
        #print(jbid,'LOADED!')
    #
    st_qcs      = state_tomagraphy_circs(order, STEPS, uniform_times=False, steps_times=best_params)
    #
    # Compute tomography fidelities for each repetition
    fids = []
    for kk in jobsd:
        fid = state_tomo(jobsd[kk].result(), st_qcs)
        fids.append(fid)
        #print(kk,'DONE')
    print('Fidelity per trial:', np.around(np.array(fids), 4))
    print()
    print('\t State tomography fidelity = {:.4f} \u00B1 {:.4f}'.format(np.mean(fids), np.std(fids)))
    print()
    print('\t Max State tomography fidelity = {:.4f}'.format(np.max(fids)))
    print()
    print('\t Expected:', np.around(result.iloc[obs]['fid_pi'],4) )
    print('-'*50)
```
